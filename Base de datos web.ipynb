```python
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html
from dash.dependencies import Input, Output

app = Dash(__name__)

colors = {
    'background': '#111111',
    'text': '#7FDBFF'
}

df = pd.read_csv('Primera_division_es.csv')
#df = df[df['Ass'] >= df['Ass'].mean()]
df = df[(df['Jugador'] != 'Total del equipo') & (df['Jugador'] != 'Total del oponente')]
#df = df[df['Edad'] <= df['Edad'].mean()]

app.layout = html.Div(style={'backgroundColor': colors['background']}, children=[
    dcc.Graph(
        id='life-exp-vs-gdp',
    ), 
    html.Br(),
    html.Label('Slider'),
    dcc.Slider(
        id='my-slider',
        min=0,
        max=int(df['Gls.'].max()),  # Asegúrate de que el valor máximo del slider no sea mayor que el número máximo de asistencias
        value=int(df['Gls.'].mean()),  # El valor inicial del slider es la media de asistencias
        marks={i: str(i) for i in range(0, int(df['Gls.'].max())+1)},  # Los marcadores del slider van desde 0 hasta el número máximo de asistencias
    ), 
    
    html.Div([
        html.Label('Dropdown'),
        dcc.Dropdown(
            id='my-dropdown',
            options=[{'label': i, 'value': i} for i in df['Equipo'].unique()],
            value=df['Equipo'].iloc[0]  # El valor inicial es el primer equipo en tu DataFrame
        ),
    ])
])

@app.callback(
    Output('life-exp-vs-gdp', 'figure'),
    [Input('my-slider', 'value'),
     Input('my-dropdown', 'value')])
def update_graph(value, selected_team):
    filtered_df = df[(df['Gls.'] >= value) & (df['Equipo'] == selected_team)]
    fig = px.scatter(filtered_df, x="Gls.", y="Ass",
                     size="Gls.", color="Jugador", hover_name="Jugador",
                     log_x=True, size_max=60)
    fig.update_layout(
        plot_bgcolor=colors['background'],
        paper_bgcolor=colors['background'],
        font_color=colors['text']
    )
    return fig

if __name__ == '__main__':
    app.run(debug=True)
```
